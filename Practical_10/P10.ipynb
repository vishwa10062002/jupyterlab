```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Covid.csv')
x=df['State']
y=df['Total Cases']
plt.figure(figsize=(10,7))
plt.xticks(rotation='vertical')
plt.xlabel('State')
plt.ylabel('Total cases')
plt.title('Total covid cases in India')
plt.bar(x,y)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Covid.csv')
df.plot.bar(x='State',y=['Total Cases','Active','Discharged','Deaths'],width=1.5,figsize=(10,7))
plt.xticks(rotation='vertical')
plt.xlabel('State')
plt.ylabel('Ratio')
plt.title('Total covid cases in India')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Covid.csv')
x=df['State']
y=df['Total Cases']
plt.pie(y,labels=x,autopct='%1.1f%%',radius=2,rotatelabels=True,pctdistance=0.9)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Covid.csv')
x=df['State']
y=df[{'Total Cases','Active','Discharged','Deaths'}]
plt.figure(figsize=(10,7))
plt.xticks(rotation='vertical')
plt.xlabel('State')
plt.ylabel('Ratio')
plt.title('Total covid cases in India')
plt.plot(x,y)
plt.legend(['Total Cases','Active','Discharged','Deaths'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Covid.csv')
x=df['State']
y=df['Total Cases']
x1=df['State']
y1=df['Active']
x2=df['State']
y2=df['Discharged']
x3=df['State']
y3=df['Deaths']
plt.figure(figsize=(10,7))
plt.xticks(rotation='vertical')
plt.xlabel('State')
plt.ylabel('Ratio')
plt.title('Total covid cases in India')
plt.scatter(x,y,marker='o',color='purple',label='Total Cases')
plt.scatter(x1,y1,marker='*',color='blue',label='Active')
plt.scatter(x2,y2,marker='^',color='red',label='Discharged')
plt.scatter(x3,y3,marker='D',color='green',label='Deaths')
plt.legend()
plt.show()
```
