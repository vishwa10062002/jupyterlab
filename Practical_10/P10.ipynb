```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Covid.csv')
x=df['State']
y=df['Total Cases']
x1=df['State']
y1=df['Active']
x2=df['State']
y2=df['Discharged']
x3=df['State']
y3=df['Deaths']
plt.figure(figsize=(10,7))
plt.xticks(rotation='vertical')
plt.xlabel('State')
plt.ylabel('Ratio')
plt.title('Total covid cases in India')
plt.scatter(x,y,marker='o',color='purple',label='Total Cases')
plt.scatter(x1,y1,marker='*',color='blue',label='Active')
plt.scatter(x2,y2,marker='^',color='red',label='Discharged')
plt.scatter(x3,y3,marker='D',color='green',label='Deaths')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Covid.csv')
x=df[{'Total Cases','Active','Discharged','Deaths','Active Ratio','Discharge Ratio','Death Ratio'}]
y=df['Deaths']
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test=train_test_split(x,y,test_size=0.3)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
LinReg=LinearRegression()
LinReg.fit(X_train, Y_train)
#Predict
y_predict=LinReg.predict(x_test)
#Error Calculation by MSE
mse=mean_squared_error(y_test, y_predict)
print('Mean Squared Error: ', mse)
#Error Calculation by RMSE
from math import sqrt
print('RMSE: ',sqrt(mse))
```
