```python
from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm
image=imread("vishwa.jpg",as_gray=True)
plt.imshow(image,cmap=cm.gray)
plt.show()
```

```python
image1=image[5:500,5:600]
plt.imshow(image1,cmap=cm.gray)
plt.show()
```

```python
image2=resize(image1,(50,60),mode='edge')
plt.imshow(image1,cmap=cm.gray)
plt.show()
print ("data type:%s,shape:%s"%(type(image2),image2.shape))
```

```python
image_row=image2.flatten()
print ("data type:%s,shape:%s"%(type(image_row),image_row.shape))
```
